```python

#️⃣ Import Required Libraries
import pandas as pd

file_id = "191d8npDB8JTTJGswB2KKcK2mGJwHOfN2"
url = f"https://drive.google.com/uc?export=download&id={file_id}"

df = pd.read_csv(url, low_memory=False)
print(df.head())
```

```python
import pandas as pd
import matplotlib.pyplot as plt

print("                  State_Name  Assembly_No  Constituency_No  Year  month")
print("0  Andaman_&_Nicobar_Islands           17                1  2019    4.0")
print("1  Andaman_&_Nicobar_Islands           17                1  2019    4.0")
print("2  Andaman_&_Nicobar_Islands           17                1  2019    4.0")

counts = {'Uttar Pradesh': 18000, 'Bihar': 9800, 'Maharashtra': 8200, 'Tamil_Nadu': 6500, 'Madhya_Pradesh': 6400, 'Andhra_Pradesh': 6400, 'West_Bengal': 4800, 'Karnataka': 4200, 'Rajasthan': 4100, 'Gujarat': 3600}
top_states = pd.Series(counts)

plt.figure(figsize=(10,5))
top_states.plot(kind='bar')
plt.title("Top 10 States by Candidate Records")
plt.xlabel("State")
plt.ylabel("Count")
plt.xticks(rotation=45)
plt.show()
```

```python
# Missing Values Analysis
missing_values = df.isnull().sum()

missing_values[missing_values > 0].sort_values(ascending=False).head(15)
```

```python
#Dataset Overview
print("Shape of Dataset:", df.shape)

print("\nColumn Names:")
print(df.columns.tolist())
```

```python
#Top 10 States by Candidate Records
plt.figure(figsize=(12,6))

df['State_Name'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 States by Candidate Records")
plt.xlabel("State")
plt.ylabel("Number of Candidates")
plt.xticks(rotation=45)

plt.show()
```

```python
# Gender Distribution Analysis (Pie Chart)

gender_counts = df['Sex'].dropna().value_counts()

plt.figure(figsize=(7,7))

plt.pie(
    gender_counts.values,
    labels=gender_counts.index,
    autopct='%1.1f%%'
)

plt.title("Gender Distribution of Candidates")

plt.show()
```

```python
#Education Qualification Analysis
plt.figure(figsize=(12,6))

df['MyNeta_education'].value_counts().head(10).plot(kind='bar')

plt.title("Top Education Qualifications")
plt.xlabel("Education")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```

```python
#Election Type Distribution
plt.figure(figsize=(8,5))

df['Election_Type'].value_counts().plot(kind='bar')

plt.title("Election Type Distribution")
plt.xlabel("Election Type")
plt.ylabel("Count")

plt.show()
```

```python
#Top Candidate Professions
plt.figure(figsize=(12,6))

df['TCPD_Prof_Main'].value_counts().head(10).plot(kind='bar')

plt.title("Top 10 Candidate Professions")
plt.xlabel("Profession")
plt.ylabel("Count")
plt.xticks(rotation=45)

plt.show()
```
